```python
plt.hist(data=df,x = 'combined_fuel_efficiency', color = 'Orange')
plt.xlabel('combined_fuel_efficiency')
plt.ylabel('frequency')
plt.title('Combined Fuel Efficiency');
```

# Bivariate Visuals
### Two Categorical Variables

```python
import numpy as np
```

```python
np.mean(df.cylinders)
```

```python
df['size'] = np.where(df.cylinders > np.mean(df.cylinders), 'large', 'small')
```

```python
df
```

```python
sns.countplot(data=df, x='year', hue= 'size')
plt.xlabel('year')
plt.ylabel('size')
plt.title('Size of Cylinders by Year');
```

### Two Quantitative Variables 

```python
plt.scatter(data=df, x = 'displacement', y = 'co2 emission')
plt.xlabel('displacement')
plt.ylabel('co2 emission')
plt.title('Displacement by CO2 Emission');
```

```python
df
```

### Categorical and Quantitative visual

```python
plt.figure(figsize=[20,5])

plt.subplot(1,2,1)
sns.violinplot(data=df, x = 'year', y='displacement')
plt.xlabel('year')
plt.ylabel('displacement')
plt.title('Displacement by Year');

plt.subplot(1,2,2)
sns.boxplot(data=df, x = 'year', y='displacement')
plt.xlabel('year')
plt.ylabel('displacement')
plt.title('Displacement by Year');
```
